```python
import numpy as np
from matplotlib import pyplot as plt
from numpy import arange
import pandas as pd

from smooth import generate_smooth_signal, generate_separated_signal
from blocks import generate_block_signal
from mixed import generated_combined_signal

FILENAME = "standard_signals.csv"
x = arange(512)
```

```python
smooth1 = generate_smooth_signal()
# This normalizes the generated signal so it always has a maximum of 1.
# This can be removed if we don't want that.
smooth1 = smooth1 / np.max(smooth1)
plt.plot(x, smooth1)
plt.ylim([0, 1])
plt.show()
```

```python
smooth2 = generate_smooth_signal()
smooth2 = smooth2 / np.max(smooth2)
```

```python
smooth3 = generate_separated_signal()
smooth3 = smooth3 / np.max(smooth3)
```

```python
smooth4 = generate_separated_signal()
smooth4 = smooth4 / np.max(smooth4)
```

```python
block1 = generate_block_signal()
block1 = block1 / np.max(block1)
```

```python
block2 = generate_block_signal()
block2 = block2 / np.max(block2)
```

```python
mix1 = generated_combined_signal()
mix1 = mix1 / np.max(mix1)
plt.plot(x, mix1)
plt.ylim([0, 1])
plt.show()
```

```python
mix2 = generated_combined_signal()
mix2 = mix2 / np.max(mix2)
plt.plot(x, mix2)
plt.ylim([0, 1])
plt.show()
```

```python
FILE = open(FILENAME, 'w')
SIGNALS = [smooth1, smooth2, smooth3, smooth4, block1, block2, mix1, mix2]
for i in x:
    FILE.write(f"{i},")
FILE.write('\n')
for signal in SIGNALS:
    for val in signal:
        FILE.write(f"{val:1.6f},")
    FILE.write('\n')
FILE.close()
```
