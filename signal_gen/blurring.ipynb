```python
import numpy as np
from matplotlib import pyplot as plt
from math import exp, sqrt, pi
from numpy.random import normal

FILENAME = "standard_signals.csv"
FILE = open(FILENAME, 'r')
for i in range(7):
    FILE.readline()
line = FILE.readline()
vals = line.split(',')
tmp = vals.pop()

for i in range(len(vals)):
    vals[i] = float(vals[i])
```

```python
k_size = 85
sig = k_size // 6
kernel = []
for i in range(k_size):
    kernel.append((1/(sig*sqrt(2*pi)) * exp(-0.5 * (i - k_size // 2)**2 / (sig**2))))
size = sum(kernel)
kernel = [k / size for k in kernel]
```

```python
blurred = np.convolve(np.asarray(vals), np.asarray(kernel), mode='same')
noisy = []
for blur in blurred:
    noisy.append(blur + normal(0, 0.02, size=None))

plt.plot(range(len(vals)), vals)
plt.plot(range(len(blurred)), blurred)
plt.plot(range(len(noisy)), noisy)
plt.legend(['GT', 'Blurred', 'Blur + Noise'])
plt.show()
```
